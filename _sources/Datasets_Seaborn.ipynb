# 5. Datasets do Seaborn

Uma das vantagens do Seaborn é que ele vem com vários conjuntos de dados embutidos, que são úteis para aprender e experimentar diferentes técnicas de visualização. Neste capítulo, exploraremos esses datasets e como acessá-los.

O Seaborn inclui vários conjuntos de dados que cobrem diferentes áreas e tipos de análise. Vamos explorar os principais:

```python
# Biblioteccas
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Listar todos os datasets disponíveis na vertical
print("Datasets disponíveis no Seaborn:")
for dataset in sns.get_dataset_names():
    print(dataset)
```

## 5.1 Dataset Iris

O dataset Iris é composto por medidas de três espécies de flores (Iris setosa, Iris versicolor e Iris virginica). Cada amostra no dataset possui as seguintes variáveis:

- sepal_length: Comprimento da sépala (em centímetros). A sépala é a parte verde que protege a flor quando ela ainda é um botão.
- sepal_width: Largura da sépala (em centímetros).
- petal_length: Comprimento da pétala (em centímetros). A pétala é a parte colorida da flor.
- petal_width: Largura da pétala (em centímetros).
- species: A espécie da flor (setosa, versicolor ou virginica). Esta é a variável que geralmente queremos prever ou entender com base nas outras medidas.

```python
# Carregar o dataset
iris = sns.load_dataset('iris')

iris
```

```python
# Visualizações
plt.figure(figsize=(15, 12))

# Boxplot do comprimento da sépala por espécie
plt.subplot(3, 2, 1)
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title('Boxplot do comprimento da sépala por espécie')

# Violin plot do comprimento da pétala por espécie
plt.subplot(3, 2, 2)
sns.violinplot(x="species", y="petal_length", data=iris, inner="quartile")
plt.title('Violin plot do comprimento da pétala por espécie')

# Swarmplot da largura da sépala por espécie
plt.subplot(3, 2, 3)
sns.swarmplot(x="species", y="sepal_width", data=iris)
plt.title('Swarmplot da largura da sépala por espécie')

# Barplot da largura média da pétala por espécie
plt.subplot(3, 2, 4)
sns.barplot(x="species", y="petal_width", data=iris, errorbar='sd')
plt.title('Barplot da largura média da pétala por espécie')

# Scatterplot: comprimento da sépala vs. comprimento da pétala
plt.subplot(3, 2, 5)
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
plt.title('Scatterplot: comprimento da sépala vs. comprimento da pétala')

plt.tight_layout()
plt.show()

```

## 5.2 Dataset Titanic

O dataset Titanic é um conjunto de dados clássico usado para aprendizado de máquina e análise exploratória de dados. Ele contém informações sobre os passageiros do navio Titanic e se eles sobreviveram ou não ao naufrágio. Aqui estão as principais variáveis:
- Survived: Indica se o passageiro sobreviveu (1) ou não (0). Esta é a variável alvo que geralmente queremos prever.
- Pclass: A classe do passageiro (1 = primeira classe, 2 = segunda classe, 3 = terceira classe).
- Sex: O sexo do passageiro (male ou female).
- Age: A idade do passageiro em anos.
- SibSp: Número de irmãos/cônjuges a bordo do Titanic.
- Parch: Número de pais/filhos a bordo do Titanic.
- Fare: A tarifa paga pelo passageiro.
- Embarked: Porto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton).
- Class: A classe do passageiro (First, Second, Third).
- Who: Indica se o passageiro é um homem, mulher ou criança.
- Adult_male: Indica se o passageiro é um homem adulto (True/False).
- Deck: O deck em que a cabine do passageiro estava localizada.
- Embark_town: A cidade de embarque do passageiro. É uma versão mais descritiva da variável "embarked".
- Alive: Indica se o passageiro sobreviveu ("yes" ou "no"). É uma versão categórica da variável "survived".
- Alone: Indica se o passageiro estava sozinho (True/False).

```python
# Carregar o dataset Titanic
titanic = sns.load_dataset('titanic')

titanic
```

```python
# Visualizações
plt.figure(figsize=(15, 10))

# Sobrevivência por classe
plt.subplot(2, 2, 1)
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Sobrevivência por Classe')

# Sobrevivência por sexo
plt.subplot(2, 2, 2)
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Sobrevivência por Sexo')

# Sobrevivência por idade
plt.subplot(2, 2, 3)
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', bins=20)
plt.title('Distribuição de Idade por Sobrevivência')

# Sobrevivência por classe e sexo
plt.subplot(2, 2, 4)
sns.barplot(x='class', y='survived', hue='sex', data=titanic)
plt.title('Sobrevivência por Classe e Sexo')

plt.tight_layout()
plt.show()
```

## 5.3 Dataset Tips

O dataset Tips contém informações sobre gorjetas em um restaurante.
- total_bill: O valor total da conta em dólares.
- tip: O valor da gorjeta em dólares.
- sex: O sexo da pessoa que pagou a conta (Male ou Female).
- smoker: Indica se a pessoa que pagou a conta é fumante (Yes ou No).
- day: O dia da semana em que a conta foi paga (Sun, Sat, Thur, Fri).
- time: A hora do dia em que a conta foi paga (Dinner ou Lunch).
- size: O número de pessoas no grupo que pagou a conta.

```python
# Carregar o dataset Tips
tips = sns.load_dataset('tips')

tips
```

```python
# Visualizações
plt.figure(figsize=(15, 10))

# Relação entre conta total e gorjeta
plt.subplot(2, 2, 1)
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
plt.title('Gorjeta vs. Conta Total por Horário')

# Gorjeta por dia da semana
plt.subplot(2, 2, 2)
sns.barplot(x='day', y='tip', data=tips)
plt.title('Gorjeta Média por Dia da Semana')

# Gorjeta por tamanho da mesa
plt.subplot(2, 2, 3)
sns.barplot(x='size', y='tip', data=tips)
plt.title('Gorjeta Média por Tamanho da Mesa')

# Distribuição de gorjetas
plt.subplot(2, 2, 4)
sns.histplot(tips['tip'], kde=True)
plt.title('Distribuição de Gorjetas')

plt.tight_layout()
plt.show()

# Relação entre gorjeta e conta total com regressão
plt.figure(figsize=(10, 6))
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Relação entre Gorjeta e Conta Total com Linha de Regressão')
plt.show()
```

## 5.4 Dataset Flights

O dataset Flights contém informações sobre o número de passageiros de voos ao longo do tempo.
Transformada em série temporal, possui a seguinte estrutura:
- year: O ano da observação. Vemos anos como 1949, 1950, 1951, 1952 e 1953 nos dados.
- month: O mês da observação (Jan, Feb, Mar, etc.). Representa todos os meses do ano.
- passengers: O número de passageiros que viajaram naquele mês e ano específicos.

```python
# Carregar o dataset Flights
flights = sns.load_dataset('flights')

# Converter para formato de série temporal
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')
print("\nDados em formato de matriz:")
print(flights_pivot.head())
```

```python
# Visualizações
plt.figure(figsize=(15, 10))

# Heatmap de passageiros por mês e ano
plt.subplot(2, 1, 1)
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Número de Passageiros por Mês e Ano')

# Linha do tempo de passageiros
plt.subplot(2, 1, 2)
sns.lineplot(x='year', y='passengers', hue='month', data=flights)
plt.title('Número de Passageiros ao Longo do Tempo por Mês')

plt.tight_layout()
plt.show()
```

## 5.5 Dataset Planets

O dataset Planets contém informações sobre planetas descobertos fora do nosso sistema solar (exoplanetas), incluindo o método de detecção, o número de planetas no sistema, o período orbital, a massa, a distância e o ano da descoberta.
- method: O método de detecção utilizado para encontrar o planeta.
- number: O número de planetas descobertos em torno da mesma estrela.
- orbital_period: O período orbital do planeta em dias.
- mass: A massa do planeta em relação à massa de Júpiter.
- distance: A distância da estrela ao planeta em parsecs.
- year: O ano em que o planeta foi descoberto.

```python
# Carregar o dataset Planets
planets = sns.load_dataset('planets')

planets
```

```python
# Visualizações
plt.figure(figsize=(15, 10))

# Contagem de planetas descobertos por método
plt.subplot(2, 2, 1)
sns.countplot(y='method', data=planets, order=planets['method'].value_counts().index)
plt.title('Número de Planetas Descobertos por Método')

# Distribuição de anos de descoberta
plt.subplot(2, 2, 2)
sns.histplot(planets['year'], bins=20, kde=True)
plt.title('Distribuição de Anos de Descoberta')

# Relação entre distância e ano de descoberta
plt.subplot(2, 2, 3)
sns.scatterplot(x='year', y='distance', hue='method', data=planets)
plt.title('Distância vs. Ano de Descoberta por Método')

# Boxplot de distância por método
plt.subplot(2, 2, 4)
sns.boxplot(x='method', y='distance', data=planets, order=planets['method'].value_counts().index[:5])
plt.title('Distribuição de Distância por Método (Top 5)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 5.6 Dataset Diamonds

O dataset diamonds do Seaborn contém informações detalhadas sobre milhares de diamantes, sendo muito utilizado para análise exploratória e visualização de dados.
- carat: Peso do diamante em quilates, que é uma das principais medidas de tamanho.
- cut: Qualidade do corte do diamante (por exemplo, Ideal, Premium, Good). O corte influencia o brilho e o valor do diamante.
- color: Cor do diamante, classificada de D (incolor) até J (mais amarelado). Quanto mais próximo de D, mais valioso geralmente é o diamante.
- clarity: Grau de pureza do diamante, indicando a quantidade de inclusões ou imperfeições internas. Exemplos: SI2, SI1, VS1, VS2.
- depth: Profundidade total do diamante como porcentagem da largura média, importante para avaliar proporções e qualidade do corte.
- table: Largura da face superior do diamante (tabela) como porcentagem da largura média, também relacionada à qualidade do corte.
- price: Preço do diamante em dólares americanos.
- x: Comprimento do diamante em milímetros.
- y: Largura do diamante em milímetros.
- z: Profundidade do diamante em milímetros.

```python
# Carregar o dataset Diamonds
diamonds = sns.load_dataset('diamonds')

diamonds
```

```python
# Visualizações
plt.figure(figsize=(12, 12))

# Distribuição de preços
plt.subplot(3, 2, 1)
sns.histplot(diamonds['price'], bins=30, kde=True)
plt.title('Distribuição de Preços')

# Preço por quilate
plt.subplot(3, 2, 2)
sns.scatterplot(x='carat', y='price', alpha=0.5, data=diamonds)
plt.title('Preço vs. Quilate')

# Preço por corte
plt.subplot(3, 2, 3)
sns.boxplot(x='cut', y='price', data=diamonds)
plt.title('Preço por Corte')

# Preço por cor
plt.subplot(3, 2, 4)
sns.boxplot(x='color', y='price', data=diamonds)
plt.title('Preço por Cor')

# Preço por claridade
plt.subplot(3, 2, 5)
sns.boxplot(x='clarity', y='price', data=diamonds)
plt.title('Preço por Claridade')

# Relação entre dimensões
plt.subplot(3, 2, 6)
sns.scatterplot(x='x', y='y', size='z', hue='price', data=diamonds, alpha=0.5)
plt.title('Dimensões do Diamante (x, y, z) e Preço')

plt.tight_layout()
plt.show()

# Relação entre quilate e preço com regressão por corte
plt.figure(figsize=(12, 8))
sns.lmplot(x='carat', y='price', hue='cut', data=diamonds.sample(1000), height=8, aspect=1.2)
plt.title('Relação entre Quilate e Preço por Corte (Amostra de 1000 diamantes)')
plt.show()
```

## 5.7 Dataset Car Crashes

O dataset car_crashes do Seaborn reúne estatísticas sobre acidentes de trânsito nos estados dos EUA, incluindo fatores de risco e custos de seguro. Veja o que cada variável representa:
- total: Número total de mortes em acidentes de trânsito por 100 mil habitantes no estado.
- speeding: Número de mortes por 100 mil habitantes relacionadas ao excesso de velocidade.
- alcohol: Número de mortes por 100 mil habitantes relacionadas ao consumo de álcool.
- not_distracted: Número de mortes por 100 mil habitantes em acidentes onde não houve distração.
- no_previous: Número de mortes por 100 mil habitantes em acidentes envolvendo motoristas sem histórico de acidentes anteriores.
- ins_premium: Valor médio anual do prêmio do seguro automotivo no estado (em dólares).
- ins_losses: Perdas médias por seguro automotivo no estado (em dólares).
- abbrev: Abreviação do nome do estado (por exemplo, AL para Alabama, AK para Alaska).

```python
# Carregar o dataset Car Crashes
car_crashes = sns.load_dataset('car_crashes')

car_crashes.head()
```

```python
# Visualizações
plt.figure(figsize=(16, 16))

# Total de acidentes por estado
plt.subplot(2, 2, 1)
sns.barplot(x='abbrev', y='total', data=car_crashes.sort_values('total', ascending=False))
plt.title('Total de Acidentes por Estado')
plt.xticks(rotation=90)

# Acidentes relacionados a álcool
plt.subplot(2, 2, 2)
sns.barplot(x='abbrev', y='alcohol', data=car_crashes.sort_values('alcohol', ascending=False))
plt.title('Acidentes Relacionados a Álcool por Estado')
plt.xticks(rotation=90)

# Correlação entre variáveis
plt.subplot(2, 2, 3)
corr = car_crashes.drop('abbrev', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlação entre Variáveis')
plt.xticks(rotation=45)

# Relação entre total de acidentes e acidentes relacionados a álcool
plt.subplot(2, 2, 4)
sns.regplot(x='total', y='alcohol', data=car_crashes)
plt.title('Relação entre Total de Acidentes e Acidentes com Álcool')

plt.tight_layout()
plt.show()
```

## 5.8 Dataset Penguins

O dataset penguins do Seaborn contém informações sobre pinguins de diferentes espécies coletados em ilhas da Antártica. Veja o que cada variável representa:
- species: Espécie do pinguim (por exemplo, Adelie, Gentoo, Chinstrap).
- island: Ilha onde o pinguim foi encontrado (por exemplo, Torgersen, Biscoe, Dream).
- bill_length_mm: Comprimento do bico do pinguim em milímetros.
- bill_depth_mm: Profundidade (altura) do bico do pinguim em milímetros.
- flipper_length_mm: Comprimento da nadadeira (asa) do pinguim em milímetros.
- body_mass_g: Massa corporal do pinguim em gramas.
- sex: Sexo do pinguim (Male ou Female).

```python
# Carregar o dataset Penguins
penguins = sns.load_dataset('penguins')

penguins.head()
```

```python
# Visualizações
plt.figure(figsize=(12, 10))

# Distribuição de comprimento de bico por espécie
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='bill_length_mm', data=penguins)
plt.title('Comprimento do Bico por Espécie')

# Distribuição de profundidade de bico por espécie
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='bill_depth_mm', data=penguins)
plt.title('Profundidade do Bico por Espécie')

# Relação entre comprimento e profundidade do bico
plt.subplot(2, 2, 3)
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=penguins)
plt.title('Comprimento vs. Profundidade do Bico por Espécie')

# Distribuição de massa corporal por espécie e sexo
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='body_mass_g', hue='sex', data=penguins)
plt.title('Massa Corporal por Espécie e Sexo')

plt.tight_layout()
plt.show()

# Pairplot para visualizar relações entre variáveis
sns.pairplot(penguins, hue='species')
plt.suptitle('Relações entre Variáveis no Dataset Penguins', y=1.02)
plt.show()
```
