# 5. Datasets do Seaborn

Uma das vantagens do Seaborn é que ele vem com vários conjuntos de dados embutidos, que são úteis para aprender e experimentar diferentes técnicas de visualização. Neste capítulo, exploraremos esses datasets e como acessá-los.

O Seaborn inclui vários conjuntos de dados que cobrem diferentes áreas e tipos de análise. Vamos explorar os principais:

```python
# Biblioteccas
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Listar todos os datasets disponíveis na vertical
print("Datasets disponíveis no Seaborn:")
for dataset in sns.get_dataset_names():
    print(dataset)
```

Vamos explorar cada um desses datasets em detalhes:

## 5.1 Dataset Iris

O dataset Iris é um dos conjuntos de dados mais famosos em estatística e aprendizado de máquina. Ele contém medidas de comprimento e largura de sépalas e pétalas de três espécies de íris.

```python
# Carregar o dataset Iris
iris = sns.load_dataset('iris')

# Visualização do dataset
plt.figure(figsize=(12, 10))

# Pairplot para visualizar relações entre variáveis
sns.pairplot(iris, hue='species', height=2.5)
plt.suptitle('Relações entre Variáveis no Dataset Iris', y=1.02)
plt.show()

# Heatmap de correlação
plt.figure(figsize=(10, 8))
corr = iris.drop('species', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação - Dataset Iris')
plt.show()
```

## 5.2 Dataset Titanic

O dataset Titanic contém informações sobre os passageiros do Titanic, incluindo se sobreviveram ou não ao naufrágio.

```python
# Carregar o dataset Titanic
titanic = sns.load_dataset('titanic')

# Visualizações
plt.figure(figsize=(15, 10))

# Sobrevivência por classe
plt.subplot(2, 2, 1)
sns.barplot(x='class', y='survived', data=titanic)
plt.title('Sobrevivência por Classe')

# Sobrevivência por sexo
plt.subplot(2, 2, 2)
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Sobrevivência por Sexo')

# Sobrevivência por idade
plt.subplot(2, 2, 3)
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', bins=20)
plt.title('Distribuição de Idade por Sobrevivência')

# Sobrevivência por classe e sexo
plt.subplot(2, 2, 4)
sns.barplot(x='class', y='survived', hue='sex', data=titanic)
plt.title('Sobrevivência por Classe e Sexo')

plt.tight_layout()
plt.show()
```

## 5.3 Dataset Tips

O dataset Tips contém informações sobre gorjetas em um restaurante.

```python
# Carregar o dataset Tips
tips = sns.load_dataset('tips')

# Visualizações
plt.figure(figsize=(15, 10))

# Relação entre conta total e gorjeta
plt.subplot(2, 2, 1)
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips)
plt.title('Gorjeta vs. Conta Total por Horário')

# Gorjeta por dia da semana
plt.subplot(2, 2, 2)
sns.barplot(x='day', y='tip', data=tips)
plt.title('Gorjeta Média por Dia da Semana')

# Gorjeta por tamanho da mesa
plt.subplot(2, 2, 3)
sns.barplot(x='size', y='tip', data=tips)
plt.title('Gorjeta Média por Tamanho da Mesa')

# Distribuição de gorjetas
plt.subplot(2, 2, 4)
sns.histplot(tips['tip'], kde=True)
plt.title('Distribuição de Gorjetas')

plt.tight_layout()
plt.show()

# Relação entre gorjeta e conta total com regressão
plt.figure(figsize=(10, 6))
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Relação entre Gorjeta e Conta Total com Linha de Regressão')
plt.show()
```

## 5.4 Dataset Flights

O dataset Flights contém informações sobre o número de passageiros de voos ao longo do tempo.

```python
# Carregar o dataset Flights
flights = sns.load_dataset('flights')

# Converter para formato de série temporal
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')
print("\nDados em formato de matriz:")
print(flights_pivot.head())

# Visualizações
plt.figure(figsize=(15, 10))

# Heatmap de passageiros por mês e ano
plt.subplot(2, 1, 1)
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Número de Passageiros por Mês e Ano')

# Linha do tempo de passageiros
plt.subplot(2, 1, 2)
sns.lineplot(x='year', y='passengers', hue='month', data=flights)
plt.title('Número de Passageiros ao Longo do Tempo por Mês')

plt.tight_layout()
plt.show()
```

## 5.5 Dataset Planets

O dataset Planets contém informações sobre planetas descobertos fora do sistema solar.

```python
# Carregar o dataset Planets
planets = sns.load_dataset('planets')

# Visualizações
plt.figure(figsize=(15, 10))

# Contagem de planetas descobertos por método
plt.subplot(2, 2, 1)
sns.countplot(y='method', data=planets, order=planets['method'].value_counts().index)
plt.title('Número de Planetas Descobertos por Método')

# Distribuição de anos de descoberta
plt.subplot(2, 2, 2)
sns.histplot(planets['year'], bins=20, kde=True)
plt.title('Distribuição de Anos de Descoberta')

# Relação entre distância e ano de descoberta
plt.subplot(2, 2, 3)
sns.scatterplot(x='year', y='distance', hue='method', data=planets)
plt.title('Distância vs. Ano de Descoberta por Método')

# Boxplot de distância por método
plt.subplot(2, 2, 4)
sns.boxplot(x='method', y='distance', data=planets, order=planets['method'].value_counts().index[:5])
plt.title('Distribuição de Distância por Método (Top 5)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 5.6 Dataset Diamonds

O dataset Diamonds contém informações sobre preços e atributos de diamantes.

```python
# Carregar o dataset Diamonds
diamonds = sns.load_dataset('diamonds')

# Visualizações
plt.figure(figsize=(12, 12))

# Distribuição de preços
plt.subplot(3, 2, 1)
sns.histplot(diamonds['price'], bins=30, kde=True)
plt.title('Distribuição de Preços')

# Preço por quilate
plt.subplot(3, 2, 2)
sns.scatterplot(x='carat', y='price', alpha=0.5, data=diamonds)
plt.title('Preço vs. Quilate')

# Preço por corte
plt.subplot(3, 2, 3)
sns.boxplot(x='cut', y='price', data=diamonds)
plt.title('Preço por Corte')

# Preço por cor
plt.subplot(3, 2, 4)
sns.boxplot(x='color', y='price', data=diamonds)
plt.title('Preço por Cor')

# Preço por claridade
plt.subplot(3, 2, 5)
sns.boxplot(x='clarity', y='price', data=diamonds)
plt.title('Preço por Claridade')

# Relação entre dimensões
plt.subplot(3, 2, 6)
sns.scatterplot(x='x', y='y', size='z', hue='price', data=diamonds, alpha=0.5)
plt.title('Dimensões do Diamante (x, y, z) e Preço')

plt.tight_layout()
plt.show()

# Relação entre quilate e preço com regressão por corte
plt.figure(figsize=(12, 8))
sns.lmplot(x='carat', y='price', hue='cut', data=diamonds.sample(1000), height=8, aspect=1.2)
plt.title('Relação entre Quilate e Preço por Corte (Amostra de 1000 diamantes)')
plt.show()
```

## 5.7 Dataset Car Crashes

O dataset Car Crashes contém informações sobre acidentes de carro nos EUA.

```python
# Carregar o dataset Car Crashes
car_crashes = sns.load_dataset('car_crashes')

# Visualizações
plt.figure(figsize=(12, 12))

# Total de acidentes por estado
plt.subplot(2, 2, 1)
sns.barplot(x='abbrev', y='total', data=car_crashes.sort_values('total', ascending=False))
plt.title('Total de Acidentes por Estado')
plt.xticks(rotation=90)

# Acidentes relacionados a álcool
plt.subplot(2, 2, 2)
sns.barplot(x='abbrev', y='alcohol', data=car_crashes.sort_values('alcohol', ascending=False))
plt.title('Acidentes Relacionados a Álcool por Estado')
plt.xticks(rotation=90)

# Correlação entre variáveis
plt.subplot(2, 2, 3)
corr = car_crashes.drop('abbrev', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlação entre Variáveis')

# Relação entre total de acidentes e acidentes relacionados a álcool
plt.subplot(2, 2, 4)
sns.regplot(x='total', y='alcohol', data=car_crashes)
plt.title('Relação entre Total de Acidentes e Acidentes com Álcool')

plt.tight_layout()
plt.show()
```

## 5.8 Dataset Penguins

O dataset Penguins contém medidas de diferentes espécies de pinguins.

```python
# Carregar o dataset Penguins
penguins = sns.load_dataset('penguins')

# Visualizações
plt.figure(figsize=(12, 10))

# Distribuição de comprimento de bico por espécie
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='bill_length_mm', data=penguins)
plt.title('Comprimento do Bico por Espécie')

# Distribuição de profundidade de bico por espécie
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='bill_depth_mm', data=penguins)
plt.title('Profundidade do Bico por Espécie')

# Relação entre comprimento e profundidade do bico
plt.subplot(2, 2, 3)
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=penguins)
plt.title('Comprimento vs. Profundidade do Bico por Espécie')

# Distribuição de massa corporal por espécie e sexo
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='body_mass_g', hue='sex', data=penguins)
plt.title('Massa Corporal por Espécie e Sexo')

plt.tight_layout()
plt.show()

# Pairplot para visualizar relações entre variáveis
sns.pairplot(penguins, hue='species')
plt.suptitle('Relações entre Variáveis no Dataset Penguins', y=1.02)
plt.show()
```
